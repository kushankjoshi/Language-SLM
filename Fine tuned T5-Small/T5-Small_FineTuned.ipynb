```python
# load the data chunk from disk to memory
from datasets import load_from_disk

# choose chunk number
choosed_chunk = 10
load_path = '/content/drive/MyDrive/'
reloaded_books_chunk = load_from_disk(load_path+'books_chunks_'+str(choosed_chunk) )
```

```python
#load tokenizer for t5-small
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")
```

```python
# configure data preprocessing pipeline
source_lang = "en"
target_lang = "hi"
prefix = "translate English to Hindi: "


def preprocess_function(examples):
    inputs = [prefix + example[source_lang] for example in examples["translation"]]
    targets = [example[target_lang] for example in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True)

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# prepare tokenized data
tokenized_books = reloaded_books_chunk.map(preprocess_function, batched=True)  # load different chunks from disk(google drive) and retarin the model
```

# Configure & train the pretrained model

## PyTorch Trainer version

```python
#Loading the t5-small model for the 1st time
from transformers import AutoModelForSeq2SeqLM
from transformers import DataCollatorForSeq2Seq


model_pytorch = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
data_collator_pytorch = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model_pytorch)
```

```python
# incase of loading model from previos checkpoint
# from transformers import AutoModelForSeq2SeqLM

!cp -r /content/drive/MyDrive/results/ /content/
loaded_model = AutoModelForSeq2SeqLM.from_pretrained('./results/checkpoint-4148')
data_collator_pytorch = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=loaded_model)
```

```python
# configure finetuning configuration
from transformers import  Seq2SeqTrainingArguments
from transformers import  Seq2SeqTrainer

output_path = save_path+"results"

training_args = Seq2SeqTrainingArguments(
    output_dir= output_path,  #'results/'
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=32, # configure batch size for train dataset
    per_device_eval_batch_size=32, # configure batch size for train dataset
    weight_decay=0.01,
    save_total_limit=2,
    num_train_epochs=1, # define epoch number
    fp16=True,
    save_strategy="steps",
    save_steps=1000,
    # load_best_model_at_end=True,
    push_to_hub=False,
)

trainer = Seq2SeqTrainer(
    model=loaded_model,  # initially model_pytorch, for from previous checkpoint loaded_model
    args=training_args,
    train_dataset=tokenized_books["train"],
    eval_dataset=tokenized_books["test"],
    tokenizer=tokenizer,
    data_collator=data_collator_pytorch,
)
```

## TensorFlow version

```python
# Load the pretrained model
from transformers import TFT5ForConditionalGeneration

model_tf = TFT5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
# define data collator for batching the data & padding
from transformers import DataCollatorForSeq2Seq

data_collator_tf = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model_tf, return_tensors="tf")
```

```python
# configure the data with batch size
tf_train_set = tokenized_books["train"].to_tf_dataset(
    columns=[ "attention_mask", "input_ids", "labels"],
    shuffle=True,
    batch_size=64, #adjust the batch size
    collate_fn=data_collator_tf,
)

tf_test_set = tokenized_books["test"].to_tf_dataset(
    columns=[ "attention_mask", "input_ids", "labels"],
    shuffle=False,
    batch_size=64, #adjust the batch size
    collate_fn=data_collator_tf,
)
```

```python
# define optimizer
from transformers import create_optimizer, AdamWeightDecay

optimizer = AdamWeightDecay(learning_rate=2e-5, weight_decay_rate=0.01)
model_tf.compile(optimizer=optimizer)
```

```python
# train(fine tune) the pretrained model
model_tf.fit(x=tf_train_set, validation_data=tf_test_set, epochs=1) #adjust the number of epochs
```
